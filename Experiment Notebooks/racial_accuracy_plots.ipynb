```python
import matplotlib.pyplot as plt

experiments = ["FER2013 only", "FER2013 + 10% EiLA", "FER2013 + 30% EiLA", "FER2013 + 50% EiLA", "FER2013 + 100% EiLA"]

black_acc = [58.9, 46.0, 43.6, 50.9, 42.3]
white_acc = [34.5, 37.6, 37.9, 37.4, 35.5]
mixed_acc = [45.3, 32.8, 31.3, 32.8, 35.9]
averages = []

for i in range(len(black_acc)):
    avg = (black_acc[i] + white_acc[i] + mixed_acc[i]) / 3
    averages.append(avg)

plt.figure(figsize=(6, 6))
plt.plot(experiments, black_acc, marker='o', label='Black', color='gold')
plt.plot(experiments, white_acc, marker='s', label='White', color='red')
plt.plot(experiments, mixed_acc, marker='^', label='Mixed', color='orange')
plt.plot(experiments, averages, marker='x', linestyle='dashed', label='Average', color='purple')

plt.xlabel('Dataset Configuration')
plt.ylabel('Accuracy (%)')
plt.title('FER2013: Accuracy Across Racial Groups')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
# Show the plot
plt.tight_layout()
plt.show()
```
