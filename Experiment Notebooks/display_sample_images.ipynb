```python
import matplotlib.pyplot as plt
import os
from PIL import Image
import IPython.display as display
import numpy as np

current_dir = os.getcwd()
dataset_dirs = [os.path.join(parent_dir, 'Experiment_Datasets/EiLA/EiLA_train'), os.path.join(parent_dir, 'Experiment_Datasets/FER2013/train_FER2013'), os.path.join(parent_dir, 'Experiment_Datasets/SFEW/Processed_Train_Faces')]

for dataset in dataset_dirs:
    train_dir = dataset
    
    valid_emotions = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
    
    def load_and_display_image(image_path):
        img = Image.open(image_path).convert('RGB')  # Ensure the image is loaded in grayscale
        img_array = np.array(img)
        img_array = 255 - img_array  # Invert the colors back if necessary
        return img_array
    
    # Calculate grid size
    num_images = len(valid_emotions)
    cols = 4  # You can adjust the number of columns
    rows = (num_images + cols - 1) // cols
    
    plt.figure(figsize=(20, 5 * rows))
    
    for i, emotion in enumerate(valid_emotions, 1):
        folder = os.path.join(train_dir, emotion)
        img_path = os.path.join(folder, os.listdir(folder)[42])
        
        img_array = load_and_display_image(img_path)
        
        plt.subplot(rows, cols, i)
        plt.imshow(img_array, cmap='gray')
        plt.title(emotion)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```
